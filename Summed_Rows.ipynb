```python
x = 15736  #skip x first rows
columns_to_keep = [1,2,3,4,5]
num1= 4
num_struct_1 = 1
num2 = num1 + 264
num_struct_2 = 132
```

```python
y= num2 #rows processed
z= 8 #rows skipped after each y rows (the gap between the states)
rows_processed = 0  # Tracks rows processed in the current `y`-row block
```

```python
 #Read the file
with open (r'output-raasi-mult1' , 'r') as f:

  #Read the file
  lines = f.readlines()
```

```python
#skip the first x rows, the split and slect specific columns
data=[line.strip().split() for line in lines[x:]] # Adjust split method if columns are not space-separated
```

```python
# Convert to DataFrame, keeping only the specified columns
df = pd.DataFrame(data)
```

```python
# Try to convert the columns to numeric, if appropriate
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
print("Length of DataFrame:\n",len(df))
```

```python
#initials of the loops
i = 0
j=0


# Store only summed rows
summed_rows_df = pd.DataFrame(columns=df.columns)
```

```python
while j < y and i < len(df):

    rows_processed=0

    #For Openshell 2
    while rows_processed < num1 and i + num_struct_1 <= len(df):

        # Calculate the sum of the current pair of rows
        row_sum = df.iloc[i:i+num_struct_1 , 1:6].sum()

        # Create a new DataFrame for the row sum with an appropriate index label
        new_index_label = f"sum_{df.index[i]}_{df.index[i+num_struct_1-1]}"
        sum_row_df = pd.DataFrame([row_sum], index=[new_index_label])

        # Append the sum row to the `summed_rows_df` DataFrame
        summed_rows_df = pd.concat([summed_rows_df, sum_row_df])

        # Split the DataFrame and insert the new row in the correct position
        df = pd.concat([df.iloc[:i+num_struct_1], sum_row_df, df.iloc[i+num_struct_1:]]) #.reset_index(drop=True) this doesn't shoe the added sum row as a new index


        # Skip the next paired rows to continue the loop correctly
        i += num_struct_1+1  # Move to the next pair after the newly added sum row

        rows_processed += num_struct_1


        #for openshell 12

    while  rows_processed < num2 and i + num_struct_2 <= len(df):
        # Calculate the sum of the current pair of rows
        row_sum = df.iloc[i:i+num_struct_2 , 1:6].sum()

        # Create a new DataFrame for the row sum with an appropriate index label
        new_index_label = f"sum_{df.index[i]}_{df.index[i+num_struct_2-1]}"
        sum_row_df = pd.DataFrame([row_sum], index=[new_index_label])

        summed_rows_df = pd.concat([summed_rows_df, sum_row_df])

        # Split the DataFrame and insert the new row in the correct position
        df = pd.concat([df.iloc[:i+num_struct_2], sum_row_df, df.iloc[i+num_struct_2:]]) #.reset_index(drop=True) this doesn't shoe the added sum row as a new index


        # Skip the next paired rows to continue the loop correctly
        i += num_struct_2+1  # Move to the next pair after the newly added sum row

        rows_processed += num_struct_2




   # After processing `y` rows, add the skipped rows directly to `summed_rows_df`
    if i + z < len(df):

        skip_rows_df = df.iloc[i:i + z, 0:5].copy()  # Capture actual skipped rows
        skip_rows_df.index = [f"skipped_{index}" for index in skip_rows_df.index]  # Rename index to indicate skipped
        summed_rows_df = pd.concat([summed_rows_df, skip_rows_df])



    i += z   #skip z rows

    j += 1

```

```python
print("Length of summed DataFrame:\n",len(summed_rows_df))
```

```python
# Output the full DataFrame and summed rows
with open('full_output.csv', 'w') as output:
    output.write("Full DataFrame with Summed Rows Added:\n\n" + df.to_string(index=True, max_rows=None, max_cols=None))

with open('summed_rows_output.csv', 'w') as output:
    output.write("Summed Rows Only:\n\n" + summed_rows_df.to_string(index=True, max_rows=None, max_cols=None))

print("Process completed. Full output and summed rows output have been written to 'full_output.csv' and 'summed_rows_output.csv'.")

```
